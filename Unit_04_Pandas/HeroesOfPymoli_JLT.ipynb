## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# find the info we want
uniqueSN = df["SN"].unique()
print(f'Total Players: {len(uniqueSN)}')

uniqueitem = df["Item ID"].unique()
print(f'Total Unique Items: {len(uniqueitem)}')

averageprice = df["Price"].mean()
print(f'Average Purchasing Price: ${round(averageprice,2)}')

totalpurchases = df["Purchase ID"].count()
print(f'Total Purchases: {totalpurchases}')

totalrevenue = df["Price"].sum()
print(f'Total Revenue: ${totalrevenue}')



# create the new dataframe for info we want
pa_df = pd.DataFrame({
    "Total Players": [len(uniqueSN)],
    "Total Unique Items": [len(uniqueitem)],
    "Average Purchasing Price": [averageprice],
    "Total Purchases": [totalpurchases],
    "Total Revenue": [totalrevenue]
})
pa_df


# First convert "average_donation", "goal", and "pledged" columns to float
# Then Format to go to two decimal places, include a dollar sign, and use comma notation

pa_df["Average Purchasing Price"] = pa_df["Average Purchasing Price"].astype(float).map("${:,.2f}".format)
pa_df["Total Revenue"] = pa_df["Total Revenue"].astype(float).map("${:,.2f}".format)

pa_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender = df[["SN", "Gender"]].drop_duplicates(keep='first')


malecount = gender[gender["Gender"] == 'Male'].count()

femalecount = gender[gender["Gender"] == 'Female'].count()

othercount = gender[gender["Gender"] == 'Other / Non-Disclosed'].count()


gender_df = pd.DataFrame({
    "Gender": ["Male", "Female", "Other / Non-Disclosed"],
    "Total Count": [malecount[0], femalecount[0], othercount[0]],
    "Percentage of Players": [(malecount[0]/len(uniqueSN)), (femalecount[0]/len(uniqueSN)), (othercount[0]/len(uniqueSN))]
})
gender_df
```

```python
gender_df.set_index("Gender")
gender_df

percent_format = {'Percentage of Players': '{:.2%}'}
gender_df.style.format(percent_format)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# average total purchase per person 
avg_df = df[["SN", "Price"]].groupby(["SN"])
avg_df_2 = avg_df.sum()
avg_df_2.head()


merged_g = pd.merge(avg_df_2, gender, on="SN", how="left")
merged_g.head()


g_price = merged_g[["Price", "Gender"]].groupby(["Gender"])
avg_ttl_price = g_price.mean()


avg_ttl_price["Price"] = avg_ttl_price["Price"] .astype(float).map("${:,.2f}".format)
avg_ttl_price
```

```python
purchase_df = df[["Gender","Price"]].groupby(["Gender"])
total_purchase_value = purchase_df.sum()
total_purchase_value

total_purchase_count = purchase_df.count()
total_purchase_count

avg_price = purchase_df.mean()
avg_price

merge_table = pd.merge(total_purchase_value, total_purchase_count, on="Gender", how="left")
merge_table_2 = pd.merge(merge_table, avg_price, on="Gender", how="left")

renamed = merge_table_2.rename(columns={"Price_x": "Total Purchase Value", "Price_y":"Total Purchase Count", "Price": "Average Purchase Price"})
renamed

merge_table_3 = pd.merge(renamed, avg_ttl_price, on="Gender", how="left")
renamed_2 = merge_table_3.rename(columns={"Price": "Average Total Purchase per Person"})
renamed_2["Average Purchase Price"] = renamed_2["Average Purchase Price"].astype(float).map("${:,.2f}".format)
renamed_2["Total Purchase Value"] = renamed_2["Total Purchase Value"].astype(float).map("${:,.2f}".format)
renamed_2

```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age = df[["SN", "Age"]].drop_duplicates(keep='first')
age

bins = [0,9,14,19,24,29,34,39,46]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "34-39", "40+"]

age["Bins"] = pd.cut(age["Age"], bins, labels=groups)
age


age_df = age[["Age", "Bins"]].groupby(["Bins"])
age_df_2 = age_df.count()
age_df_2


age_df_2["Percentage of Players"] = age_df_2["Age"]/len(uniqueSN)
age_df_2

percent_format = {'Percentage of Players': '{:.2%}'}
age_df_2.style.format(percent_format)

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_p = df[["Age", "Price"]]
age_p

bins = [0,9,14,19,24,29,34,39,46]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "34-39", "40+"]

age_p["Bins"] = pd.cut(age_p["Age"], bins, labels=groups)
age_p


age_p_bins = age_p[["Bins", "Price"]].groupby(["Bins"])
average_price_per_bin = age_p_bins.mean()
average_price_per_bin

average_purchase_count_b = age_p_bins.count()
average_purchase_count_b

total_purchase_value_b = age_p_bins.sum()
total_purchase_value_b

merge_table_b = pd.merge(average_price_per_bin, average_purchase_count_b, on="Bins", how="left")
merge_table_b_2 = pd.merge(merge_table_b, total_purchase_value_b, on="Bins", how="left")
merge_table_b_2

renamed_b = merge_table_b_2.rename(columns={"Price_x": "Average Purchase Price", "Price_y":"Purchase Count", "Price": "Total Purchase Value"})
renamed_b["Average Purchase Price"] = renamed_b["Average Purchase Price"].astype(float).map("${:,.2f}".format)
renamed_b["Total Purchase Value"] = renamed_b["Total Purchase Value"].astype(float).map("${:,.2f}".format)
renamed_b

```

```python

avg_df = df[["SN", "Price"]].groupby(["SN"])
avg_df_2 = avg_df.sum()
avg_df_2.head()

merged_b = pd.merge(avg_df_2, age, on="SN", how="left")
merged_b.head()

b_price = merged_b[["Price", "Bins"]].groupby(["Bins"])
avg_ttl_b = b_price.mean()
avg_ttl_b



age_bins_f = pd.merge(renamed_b, avg_ttl_b, on="Bins", how="left")
final_age = age_bins_f.rename(columns={"Price":"Average Total Purchase per Person"})
final_age["Average Total Purchase per Person"] = final_age["Average Total Purchase per Person"].astype(float).map("${:,.2f}".format)
final_age
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = df[["Item ID", "Item Name", "Price"]]

ttlitems = items[["Item ID", "Price"]].groupby(["Item ID"])
totalitemcount = ttlitems.count()
totalitemcount

totalitemvalue = ttlitems.sum()
totalitemvalue.sort_values(by="Price", ascending=False)

mergeitems = pd.merge(totalitemcount, totalitemvalue, on="Item ID", how="left")
mergeitems = mergeitems.rename(columns={"Price_y": "Total Purchase Value", "Price_x": "Purchase Count"})
mergeitems

# mergeitems2 = pd.merge(mergeitems, ttlitems.value(), on="Item ID", how="left")
# mergeitems2
iteminfo = items[["Item ID", "Item Name", "Price"]].drop_duplicates(keep='first')
mergeitems_2 = pd.merge(mergeitems, iteminfo, on="Item ID", how="left")




mergeitems_2["Price"] = mergeitems_2["Price"].astype(float).map("${:,.2f}".format)
mergeitems_2.sort_values(by="Purchase Count", ascending=False).head()


```
